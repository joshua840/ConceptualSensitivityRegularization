```python
import pandas as pd
import yaml

path = '/media/disk1/Data/SpuriousCatDog/metadata.csv'
df = pd.read_csv(path)

# shuffle
df_tr = df.query('split == "train"').sample(frac=1).reset_index(drop=True)

concept_dict = {
    "water": df_tr.query('place == 0 and y == 0').img_filename.tolist()[:32] + df_tr.query('place == 0 and y == 1').img_filename.tolist()[:32],
    "land": df_tr.query('place == 1 and y == 0').img_filename.tolist()[:32] + df_tr.query('place == 1 and y == 1').img_filename.tolist()[:32],
}

with open('configs/dataset/catdog_concepts_v2.yaml', 'w') as file:
    yaml.dump(concept_dict, file)
```
