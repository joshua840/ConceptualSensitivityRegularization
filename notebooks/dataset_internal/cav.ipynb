```python
import os
import torch
from argparse import Namespace
from csr.module.dataset.feature_data_module import EpochChangeableFeatureDataset
from csr.module.utils.cav import compute_cav

args = {
    "dataset": "catdog",
    "model_name": "convnext_t",
    "cav_type": "svm",
    "root": "/media/disk2/Data",
    "save_path": "./data/cav"
}

args = Namespace(**args)

dataset = EpochChangeableFeatureDataset(
    split="tr",
    root=args.root,
    dataset=args.dataset,
    model_name=args.model_name,
)

x = dataset.x
y = dataset.y

w = compute_cav(
    vecs=x,
    targets=y,
    cav_type=args.cav_type,
)

# save
torch.save(
    w,
    os.path.join(
        args.save_path, f"{args.dataset}_{args.model_name}_{args.cav_type}.pt"
    ),
)
```
